```python
import numpy as np
from scipy import stats
import seaborn as sns

%matplotlib inline
```

```python
data = [30,36,47,50,52,52,56,60,63,70,70,110]
```

```python
# mean
np.mean(data)
```

```python
# median
np.median(data)
```

```python
# mode
stats.mode(data)
```

```python
# variance
np.var(data)
```

```python
# 25 percentile
np.percentile(data, 25)
```

```python
# 75 percentile
np.percentile(data, 75)
```

```python
# interquantile range
stats.iqr(data)
```

```python
sns.boxplot(data)
```

```python
sns.violinplot(data)
```

```python
sns.distplot(data)
```

```python
sns.distplot(data, hist=False, rug=True)
```

```python
sns.kdeplot(data, shade=True)
```
